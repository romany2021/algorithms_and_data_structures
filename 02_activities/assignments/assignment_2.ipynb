#### Starter Code for Question 2

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val = 0, left = None, right = None):
#         self.val = val
#         self.left = left
#         self.right = right

from typing import List

class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def bt_path(root: TreeNode) -> List[List[int]]:
    paths = []

    def dfs(node, current_path):
        if node is None:
            return
        current_path = current_path + [node.val]
        if node.left is None and node.right is None:
            paths.append(current_path)
        else:
            dfs(node.left, current_path)
            dfs(node.right, current_path)

    dfs(root, [])
    return paths
```

```python
root1 = TreeNode(1,
            TreeNode(2, TreeNode(3), TreeNode(5)),
            TreeNode(2, TreeNode(6), TreeNode(7)))
print(bt_path(root1))

root2 = TreeNode(10,
            TreeNode(9, TreeNode(8)),
            TreeNode(7))
print(bt_path(root2))
```

```python
# Your answer here
# New example:
# Input:  nums = [8, 3, 5, 3, 8]
# Output: 3
# Why: scanning left to right, 3 is the first value seen twice (the second 3 at index 3).

# nums = [4, 5, 6, 4, 6]
# x=4 -> not in seen -> add 4 -> seen = {4}
# x=5 -> not in seen -> add 5 -> seen = {4,5}
# x=6 -> not in seen -> add 6 -> seen = {4,5,6}
# x=4 -> already in seen -> return 4 <- stops here
```


-   Copy the solution your partner wrote. 


```python
# Your answer here

from typing import List

def first_duplicate(nums: List[int]) -> int:
    seen = set()
    
    for num in nums:
        if num in seen:
            return num
        seen.add(num)
    
    return -1

nums_1 = [3, 1, 4, 2, 5, 1, 6]
nums_2 = [7, 8, 9, 10]
nums_3 = [4, 5, 6, 4, 6]

print(first_duplicate(nums_1))
print(first_duplicate(nums_2))
print(first_duplicate(nums_3))
```


-   Explain why their solution works in your own words.


```python
# Your answer here
# 0(1) is being used as primary approach to checking memberships. This works well as in this setup it doesn't slow down the execution.
# What the code does is walking left to right and checks if it has seen the value before, and if it sees the same vaule again, it immeidately stops the code as it has identified the duplicate.
# This also guarantees return on the first hit since code is being stopped the moment it catches the duplicate, and if no duplicate found it know to return -1.
```


-   Explain the problem’s time and space complexity in your own words.


```python
# Your answer here
#Time complexity: O(n). The loop visits each of the n values once, and each set membership check ("x in seen") and insertion ("seen.add") is O(1) on average. n items x O(1) work each = O(n).
# Space complexity: O(n) may lead to situation when every value gets stored in "seen", so the set grows to the full size of the list, n items and if there is no duplicate it will take space without results.
```
